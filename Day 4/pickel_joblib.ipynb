```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import pickle
import joblib
```

```python
df = pd.read_csv("datasets/coimbatore_pois_features_500m_clean.csv")
df.head()
```

```python
feature_cols = [
    "services_500m",
    "transport_500m",
    "roads_500m",
    "major_roads_500m",
    "landuse_residential_500m",
    "landuse_commercial_500m",
    "landuse_industrial_500m",
    "landuse_retail_500m",
    "landuse_mixed_500m",
    "landuse_farmland_500m",
    "landuse_other_500m",
]
target_col = "pois_500m"

model_df = df[feature_cols + [target_col]].dropna()
X = model_df[feature_cols]
y = model_df[target_col]
```

```python
model = linear_model.LinearRegression()
model.fit(X, y)
model.coef_
model.intercept_
```

```python
with open("model_pickle.pkl", "wb") as file:
    pickle.dump(model, file)

with open("model_pickle.pkl", "rb") as file:
    mp = pickle.load(file)

mp.coef_
mp.intercept_
```

```python
example_input = X.iloc[[0]]
mp.predict(example_input)
```

```python
joblib.dump(model, "model_joblib.pkl")
mj = joblib.load("model_joblib.pkl")

mj.coef_
mj.intercept_
mj.predict(example_input)
```
