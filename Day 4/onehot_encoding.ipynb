```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(
    "datasets/dataset_final.csv",
    engine="python",
    on_bad_lines="skip",
 )
df.head()
```

```python
target_col = "pois_500m"
feature_cols = [
    "services_500m",
    "transport_500m",
    "roads_500m",
    "major_roads_500m",
    "landuse_residential_500m",
    "landuse_commercial_500m",
    "landuse_industrial_500m",
    "landuse_retail_500m",
    "landuse_mixed_500m",
    "landuse_farmland_500m",
    "landuse_other_500m",
]

missing = [c for c in feature_cols + [target_col] if c not in df.columns]
if missing:
    raise ValueError(f"Missing columns: {missing}")

X = df[feature_cols]
y = df[target_col]
```

```python
model = LinearRegression()
model.fit(X, y)
model.coef_
model.intercept_
model.score(X, y)
# Large dataset: show only the first 5 predictions
model.predict(X.head())
```

```python
example_input = X.iloc[[0]]
model.predict(example_input)

another_input = X.iloc[[1]]
model.predict(another_input)
```
